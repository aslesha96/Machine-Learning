```python
covar_matrix = np.matmul(datastd.T,datastd)
covar_matrix.shape
```

```python
from scipy.linalg import eigh 
values, vectors = eigh(covar_matrix, eigvals=(588,589))
print("Shape of eigen vectors = ",vectors.shape)
vectors = vectors.T
print("Updated shape of eigen vectors = ",vectors.shape)
```

```python
import matplotlib.pyplot as plt
new_coordinates = np.matmul(vectors, datastd.T)

print (" resultanat new data points' shape ", vectors.shape, "X", datastd.T.shape," = ", new_coordinates.shape)
```

```python
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
pca.n_components = 590
pca_data = pca.fit_transform(datastd)
# print("shape of pca_reduced.shape = ", pca_data.shape)
percentage_var_explained = pca.explained_variance_ / np.sum(pca.explained_variance_);
cum_var_explained = np.cumsum(percentage_var_explained)
plt.figure(1, figsize=(6, 4))

plt.clf()
plt.plot(cum_var_explained, linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n_components')
plt.ylabel('Cumulative_explained_variance')
plt.show()
```
